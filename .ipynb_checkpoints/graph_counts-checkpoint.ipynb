```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
counts = {"Exists Temporal Localization": 710734, "Interaction Temporal Localization": 263784, "Choose": 185343, "Equals": 169958, "First/Last": 101091, "Interaction": 74515, "Conjunction": 31534, "Relation Exists": 28364, "Object Exists": 15539, "Longest/Shortest Action": 460}
keys = list(counts.keys())
values = []
for key in keys:
    values.append(counts[key])
total = np.sum(values)
```

```python
size = 5
hfont = {'fontname':'Verdana'}
color_idx = color_idx=[0, 2, 4, 6, 8, 10, 11, 9, 5]
cmap = plt.get_cmap('Paired')
init = 0
step = 2
c = cmap(np.array(color_idx))
sns.set_style('whitegrid')
plt.subplots(figsize=(size*8, size*len(counts)))
ax = sns.barplot(x=values, y=keys, palette=c)
plt.title('Question Type', fontsize=size*55, **hfont)
plt.xlabel('Count (100K)', fontsize=size*40, **hfont)
plt.xticks(fontsize=size*35, **hfont)
plt.yticks(fontsize=size*35, **hfont)
i = 0
space=0.4
sz=30
sns.despine()
xlabels = [int(x) for x in ax.get_xticks()/100000]
#print(xs)
#xlabels = ['{:,.2f}'.format(x) for x in xs]

ax.set_xticklabels(xlabels)
for p in ax.patches:
    _x = p.get_x() + p.get_width() + float(space)
    _y =  p.get_y() + p.get_height() *  1/2
    key = keys[i]
    value = str(round(counts[key]/total * 100, 1)) + '%'
    ax.text(_x, _y, value, ha="left", va="center",fontsize=sz*4)
    i += 1
plt.show()
```
