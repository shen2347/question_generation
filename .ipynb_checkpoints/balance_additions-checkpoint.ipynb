```python
import json
import pickle
import random
import pandas as pd
```

```python
#infile = open('../data/test_videos_stsg.pkl', 'rb')
infile = open('../data/training_camera_ready_stsgs.pkl', 'rb')
train_stsgs = pickle.load(infile)
infile.close()
```

```python
path = '../../../Desktop/final_balanced'
```

```python
vs = set()
```

```python
df = pd.DataFrame()
dfs = []
for i, v_id in enumerate(train_stsgs):
    if i % 1000 == 0:
        print(i)
    try:
        with open('%s/%s.json' % (path, v_id), 'rb') as f:
            QA = json.load(f)
    except:
        vs.add(v_id)
        continue
    
    dicts = []
    for q_id in QA:
        q = QA[q_id]
        q['program'] = q_id
        q['first_word'] = q['program'].split('(')[0]
        q['video'] = v_id
        if q['answer'] == None:
            q['none'] = 1
        else:
            q['none'] = 0
            
        q['yes/no'] = q['answer'] in ['Yes', 'No']
        dicts.append(q)
    
    video_df = pd.DataFrame(dicts)
    dfs.append(video_df)
print(len(dfs))
df = pd.concat(dfs)
```

```python
by_p = {}

for i in dicts:
    p = i['program']
    a = i['answer']
    
    if p not in by_p:
        by_p[p] = {
            'Yes': [],
            'No': []
        }
        
    by_p[p][a].append(i)
        
    
```

```python
len(by_p)
```

```python
keep_qs = []
for p in by_p:
    ans = by_p[p]
    num_yes = len(ans['Yes'])
    num_no = len(ans['No'])
    if num_yes == 0 or num_no == 0:
        print('kept wrong!', p)
        
    num_keep = min(num_yes, num_no)
    keep_qs += random.sample(ans['Yes'], k=num_keep)
    keep_qs += random.sample(ans['No'], k=num_keep)
    
    
```

```python
len(keep_qs)
```

```python
by_p2 = {}

for i in keep_qs:
    p = i['program']
    a = i['answer']
    
    if p not in by_p2:
        by_p2[p] = {
            'Yes': [],
            'No': []
        }
        
    by_p2[p][a].append(i)
        
    
```

```python
ans = 0
no_ans = 0
none = 0
answers = {}
for i, v_id in enumerate(train_stsgs):
    if i % 1000 == 0:
        print(i)
    try:
        with open('%s/%s.json' % (path, v_id), 'rb') as f:
            QA = json.load(f)
    except:
        vs.add(v_id)
        continue
    
    for q_id in QA:
        q = QA[q_id]
        if 'answer' in q:
            a = q['answer']
            if a not in answers:
                answers[a] = 0
            answers[a] += 1
            ans += 1
            
            if a == None:
                none += 1
        else:
            no_ans += 1
            
ans, no_ans
```
