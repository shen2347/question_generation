```python
def replaceFirstIDAndMakeLast(QA, changed):
    ids = list(QA.keys())
    first_id = ids[0]
    last_id = ids[-1]
    
    v_id, i = last_id.split('-')
    i = int(i)
    
    new_id = '%s-%s' %(v_id, i+1)
    
    QA[new_id] = QA[first_id]
    del QA[first_id]
    
    return QA, changed
```
