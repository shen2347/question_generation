```python
hme_score =[]
hme_accuracy= []

hme_root_accuracy = []

dicts_hme = []

rule = 'Overall'

consistency_rule_list = ['Interaction', 'After', 'Before', 'While', 'Between',
                         'And', 'Xor', 'First', 'Last', 'Equals', 'Choose',
                         'Overall']





for rule in consistency_rule_list:
    for key, x in hierarchy_hme.items():
        acc, root_acc = x
        
        acc *= 100
        if key not in consistency_hme:
            continue

        rule_dicts = consistency_hme[key]
        if rule not in rule_dicts:
            continue
        score = rule_dicts[rule]["Score"]

        hme_accuracy.append(acc)
        hme_root_accuracy.append(root_acc)
        hme_score.append(score)
        d = {'model': 'HME', 'Consistency': score, 'Accuracy': acc, 'Root_acc': root_acc}

        dicts_hme.append(d)

        
    print(round(sum(hme_score) / len(hme_score) * 100, 2), rule)
```

```python
hcrn_score =[]
hcrn_accuracy= []
hcrn_root_accuracy = []

dicts_hcrn = []

rule = 'Overall'

consistency_rule_list = ['Interaction', 'After', 'Before', 'While', 'Between',
                         'And', 'Xor', 'First', 'Last', 'Equals', 'Choose',
                         'Overall']


for rule in consistency_rule_list:
    for key, x in hierarchy_hcrn.items():
        acc, root_acc = x
        acc *= 100
        if key not in consistency_hcrn:
            continue

        rule_dicts = consistency_hcrn[key]
        if rule not in rule_dicts:
            continue
        score = rule_dicts[rule]["Score"]

        hcrn_accuracy.append(acc)
        hcrn_score.append(score)
        hcrn_root_accuracy.append(root_acc)
        d = {'model': 'HCRN', 'Consistency': score, 'Accuracy': acc, 'Root_acc': root_acc}

        dicts_hcrn.append(d)
        
    print(round(sum(hcrn_score) / len(hcrn_score) * 100, 2), rule)
```

```python
ml_score =[]
ml_accuracy= []
ml_root_accuracy = []

dicts_ml = []

rule = 'Overall'

consistency_rule_list = ['Interaction', 'After', 'Before', 'While', 'Between',
                         'And', 'Xor', 'First', 'Last', 'Equals', 'Choose',
                         'Overall']


for rule in consistency_rule_list:
    for key, x in hierarchy_ml.items():
        acc, root_acc = x
        acc *= 100
        if key not in consistency_ml:
            continue

        rule_dicts = consistency_ml[key]
        if rule not in rule_dicts:
            continue
        score = rule_dicts[rule]["Score"]

        ml_accuracy.append(acc)
        ml_score.append(score)
        ml_root_accuracy.append(root_acc)
        d = {'model': 'Most Likely', 'Consistency': score, 'Accuracy': acc, 'Root_acc': root_acc}

        dicts_ml.append(d)
        
    print(round(sum(ml_score) / len(ml_score) * 100, 2), rule)
```

```python
model_name = 'most_common'
folder = f'analysis_results/seen_{model_name}'
files = os.listdir(folder)

by_tp_hme = {}

for num, file in enumerate(files):
    try:
        with open(f'{folder}/{file}', 'rb') as f:
            data = pickle.load(f)
    except:
        continue
        
    for q_tp in data:
        trans_q_tp = collapsed_qtypes[q_tp]
        if trans_q_tp not in by_tp_hme:
            by_tp_hme[trans_q_tp] = {"Wrong": 0, "Total": 0}
        t = data[q_tp]
        for ques in t:
            for composition in t[ques]:
                if composition == 'parent_correct':
                    continue
                    
                
                    
                by_tp_hme[trans_q_tp]['Wrong'] += t[ques][composition]['Wrong']
                by_tp_hme[trans_q_tp]['Total'] += t[ques][composition]['Total']
                 
                
                
by_tp_hme
```

```python
tot = 0
tot_w = 0

for i in by_tp_hme:
    x = by_tp_hme[i]
    t = x['Total']
    w = x['Wrong']
    tot += t
    tot_w += w
    
    if t == 0:
        print("N/A %s" % i)
        continue
    perc = round((t-w) / t * 100, 2)
    print(perc, i)
```

```python
round((tot - tot_w) / tot * 100, 2)
```

# Binary compositional reasoning

1. replace compositionality scores with either 0 if not already 1, or 1
2. Use whether the top level question is correct


```python
def bny(score):
    if score == 1:
        return score
    else: 
        return 0
```

```python
# Peason coefficient


# seed random number generator
seed(1)
# prepare data
data1 = [bny(i) for i in hme_score]
data2 = hme_root_accuracy

# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)
print('HME Pearsons correlation: %.3f' % corr)


data1 = [bny(i) for i in hcrn_score]
data2 = hcrn_root_accuracy

# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)
print('HCRN Pearsons correlation: %.3f' % corr)
```

### Try not heirarchy-wide

```python
model_name = 'hme'
folder = f'analysis_results/seen_{model_name}'
files = os.listdir(folder)

bny_score = []
parent_acc = []

for num, file in enumerate(files):
    try:
        with open(f'{folder}/{file}', 'rb') as f:
            data = pickle.load(f)
    except:
        continue
        
    for q_type in data:
        for ques in data[q_type]:
            parent_correct = data[q_type][ques]['parent_correct']
            for composition in data[q_type][ques]:
                if composition == 'parent_correct':
                    continue
                    
                if parent_correct:
                    parent_acc.append(1)
                else:
                    parent_acc.append(0)
                    
                if data[q_type][ques][composition]['Wrong'] == 0:
                    bny_score.append(1)
                else:
                    bny_score.append(0)
                
                
        
```

```python
len(bny_score), len(parent_acc)
```

```python


data1 = bny_score
data2 = parent_acc

# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)
print('HCRN Pearsons correlation: %.3f' % corr)
```
