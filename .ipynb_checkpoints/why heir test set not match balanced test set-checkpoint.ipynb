```python
with open('../../../Desktop/extra_keys.pkl', 'rb') as f:
    extra_keys = pickle.load(f)
```

```python
extra_keys = list(extra_keys)
```

```python
by_v = {}
for k in extra_keys:
    v = k[:5]
    if v not in by_v:
        by_v[v] = []
    by_v[v].append(k)
```

```python
qs = []

for v in by_v:
    with open('../exports/dataset/balanced/test/%s.txt' % v, 'r') as f:
        QA = json.load(f)
        
    
    for q_id in by_v[v]:
        new_q = {}
        
        q = QA[q_id]
        
        new_q['type'] = q['attributes']['type']
        new_q['answer'] = q['answer']
        new_q['video'] = v
        qs.append(new_q)
```

```python
for v in by_v:
    if len(by_v[v]) != len(set(by_v[v])):
        print('weweww')
        break
```
