# Get lists of video_ids

```python
def getSTSGS(group):
    if group == 'train': 
        return train_stsgs
    elif group == 'test':
        return test_stsgs
    else:
        print('invalid group', group)
        return

def save_urls(group, start, stop, video_folder):
    stsgs = list(getSTSGS(group).keys())
    url_path = "https://visualgenome.org/hits/agqa_videos?filename="
    src_path = '../Charades_v1_480/%s.mp4'
    dest_path = '../AMT/easyturk/easyturk/videos/%s/%s.mp4'
    
    video_ids = stsgs[start:stop]
    amt_input = []
    
    for v in video_ids:
        # add video url path
        url = url_path + v
        amt_input.append({'url': url})
        
        # get videos to upload to scene graphs
        src_file = src_path % v
        dest_file = dest_path % (video_folder, v)
        shutil.copyfile(src_file, dest_file)
        
    with open('../AMT/easyturk/easyturk/data/object_dataset/%s-%s.txt' % (start, stop), 'w+') as f:
        json.dump(amt_input, f)
        
    return amt_input

    


amt_input = save_urls('train', 0, 200, '11-12')
```

# Process results

```python
import pickle
import pandas as pd
```

```python
def loadResults(paths):
    results = {}
    for p in paths:
        with open('%s.txt' % p, 'rb') as infile:
            new_results = pickle.load(infile)
        results.update(new_results)
        
    dicts = []
    zipped = {}
    by_obj = []
    for i in results:
        for j in results[i]:
            worker = j['worker_id']
            for k in j['output']:
                k['worker'] = worker
            #dicts += j['output']
            
            for d in j['output']:
                exists = d['objNoExist'][0]
                objs = d['objs']
                
                zipped[d['url'][-5:]] = list(zip(objs, exists))
                
                new_d = {'url': d['url']}
                for idx in range(len(exists)):
                    new_d[objs[idx]] = exists[idx]
                
                dicts.append(new_d)
            
    df = pd.DataFrame(dicts)
    
    return df, dicts, zipped

df, dicts, zipped = loadResults(['test_objects_me'])
```

# Getting questions from AMT results

```python
# just an example, there is a file where you can find the actions for all the actions
actions_list = {"H46LQ" :['putting a box somewhere', 'taking a bag from somewhere', 'closing a laptop', 'holding some food', 'holding a phone']}


```

```python
# this is to get the articles
objArticle = {"person": "a", "bag": "a", "bed": "a", "blanket": "a", "book": "a", "box": "a", "broom": "a", "chair": "a", "closet/cabinet": "a", "clothes": "some", 
    "cup/glass/bottle": "a", "dish": "a", "door": "a", "doorknob": "a", "doorway": "a", "floor": "the", "food": "some", "groceries": "some", "laptop": "a", 
    "light": "a", "medicine": "a", "mirror": "a", "paper/notebook": "a", "phone/camera": "a", "picture": "a", "pillow": "a", "refrigerator": "a", "sandwich": "a", 
    "shelf": "a", "shoe": "a", "sofa/couch": "a", "table": "a", "television": "a", "towel": "a", "vacuum": "a", "window": "a", "hands": "their", "hair": "their", 
    "closet": "a", "cup": "a", "paper": "a", "phone": "a", "sofa": "a", "other person":"some"
}
```

```python
questions = []

for dict in dicts:
    vid_id = dict['url'][-5:]
    No_objs = []
    for obj, value in dict.items():
        if value == True:
            No_objs.append(obj)
            article = objArticle[obj]
            question = f'Does {article} {obj} exist?'
            questions.append(question)
            if vid_id in actions_list:
                actions = actions_list[vid_id]
                for i, action1 in enumerate(actions):
                    questions.append(f'{question[:-1]} after {action1}?')
                    questions.append(f'{question[:-1]} before {action1}?')
                    questions.append(f'{question[:-1]} while {action1}?')
                    for action2 in actions[i+1:]:
                        questions.append(f'{question[:-1]} before {action1} but after {action2}?')
                        questions.append(f'{question[:-1]} before {action2} but after {action1}?')
```
