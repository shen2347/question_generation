```python
import json
import random
import balance
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import grammar as g
```

```python

#infile = open('../data/test_videos_stsg.pkl', 'rb')
infile = open('../data/stsgs/test_stsgs.pkl', 'rb')
test_stsgs = pickle.load(infile)
infile.close()

#infile = open('../data/videos_stsg.pkl', 'rb')
infile = open('../data/stsgs/train_stsgs.pkl', 'rb')
train_stsgs = pickle.load(infile)
infile.close()
```

```python

def getQA(bal, group, v_id):
    with open('../exports/dataset/%s/%s/%s.txt' % (bal, group, v_id), 'rb') as f:
        test_qs = json.load(f)
    return test_qs
```

# Train

```python
all_train_ids = []
for i, v_id in enumerate(train_stsgs): 
    if i % 50 == 0:
        print(i)
    #with open('../exports/dataset/json_format_all/train/%s.txt' % (v_id), 'rb') as f:
    #    test_qs = list(json.load(f))
        
    
    with open('../exports/dataset/all/train/%s.txt' % (v_id), 'rb') as f:
        test_qs = list(json.load(f))
        
    all_train_ids.append(test_qs)
```

```python
all_train_ids2 = []

for i in all_train_ids:
    all_train_ids2 += i
```

```python
with open('../exports/dataset/json_format_all/train/train_q_ids.txt', 'w+') as f:
    json.dump(all_train_ids2, f)
        
```

```python
#train_indexes = random.sample(range(1, 192000000), 2000000)
train_indexes = random.sample(all_train_ids2, 2000000)
```

```python
with open('../exports/dataset/json_format_all/train/unbalanced_subset_train_q_ids.txt', 'w+') as f:
    json.dump(train_indexes, f)
```

### Get the questions

```python
by_vid_tr = {}
for q_id in train_indexes:
    v = q_id[:5]
    if v not in by_vid_tr:
        by_vid_tr[v] = []
        
    by_vid_tr[v].append(q_id)
```

```python
train_sampled_q = []
for i, v_id in enumerate(by_vid_tr):
    if i % 10 == 0:
        print(i)
    with open('../exports/dataset/json_format_all/train/%s.txt' % (v_id), 'rb') as f:
        QA = json.load(f)
        
    for q_id in by_vid_tr[v_id]:
        
        train_sampled_q.append(QA[q_id])
    
```

```python
with open('../exports/dataset/json_format_all/train_unbalanced_subset.txt', 'w+') as f:
    json.dump(train_sampled_q, f)
```

# Test

```python
all_test_ids = []
for i, v_id in enumerate(test_stsgs): 
    if i % 50 == 0:
        print(i)
    #with open('../exports/dataset/json_format_all/test/%s.txt' % (v_id), 'rb') as f:
    #    test_qs = list(json.load(f))
        
    
    with open('../exports/dataset/all/test/%s.txt' % (v_id), 'rb') as f:
        test_qs = list(json.load(f))
        
    all_test_ids.append(test_qs)
```

```python
all_test_ids2 = []

for i in all_test_ids:
    all_test_ids2 += i
```

```python
with open('../exports/dataset/json_format_all/test/test_q_ids.txt', 'w+') as f:
    json.dump(all_test_ids2, f)
        
```

```python

test_indexes = random.sample(all_test_ids2, 1000000)
```

```python
with open('../exports/dataset/json_format_all/test/unbalanced_subset_test_q_ids.txt', 'w+') as f:
    json.dump(test_indexes, f)
```
