```python
def loadResults(paths):
    results = {}
    for p in paths:
        with open('../AMT/easyturk/easyturk/results/%s.txt' % p, 'rb') as infile:
            new_results = pickle.load(infile)
        results.update(new_results)
        
    dicts = []
    for i in results:
        for j in results[i]:
            worker = j['worker_id']
            for k in j['output']:
                k['worker'] = worker
                
                if getTemplTranslation(k['template']) == "Interaction" or k['template'] == 'Interaction':
                    for word in ['before', 'after', 'while', 'between', 'Before', 'After', 'While', 'Between']:
                        if word in k['question']:
                            k['template'] = 'Interaction Temporal Localization'
                    else:
                        print(k['question'])
                    
            dicts += j['output']
            

    df = pd.DataFrame(dicts)
    df['correct'] = df['closest_match'].apply(lambda x: 1 if x == 'Correct' else 0)
    df['AGQA_question'] = df['template'].apply(lambda x: 1 if x in AGQA_templ else 0)
    df['type'] = df['template'].apply(getTemplTranslation)
    #df['type'] = df['type1'].apply(getTemplTranslation2)
    df['replacement_answer'] = df['free_form_answer'].apply(lambda x: 1 if len(x) > 0 else 0)
    
    data = df[df['ans_replaced'] == False]
    ans_rep = df[df['ans_replaced']]
    
    return df, data, ans_rep
```

```python
paths = [
    'results-155',
    'results-first4',
    'results-fill10-all',
    'results-fill40-first36',
    'results-fill25-first24',
    'results-fill25-second2',
]

paths = [
    'results-155',
    'results-first4',
    'results-fill10-all',
    'results-fill40-first37',
    'results-fill25-first25',
    'results-fill25-second2',
]



df, data, ans_rep = loadResults(paths)
```

```python
def accPerAns(df, threshold=3):
    vc = df['gtanswer'].value_counts()
    vals_to_remove = vc[vc < threshold].index.values
    df['too_low_ans'] = df['gtanswer'].apply(lambda x: x in vals_to_remove)
    df = df[df['too_low_ans'] == False]
    
    x = df.groupby('gtanswer').mean()['correct'].to_frame()
    x = x.sort_values('correct')
    return x
    
accPerAns(data, threshold=4)
```
