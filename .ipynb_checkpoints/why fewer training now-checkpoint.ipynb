```python

with open('../../../Desktop/top_keys (1).pkl', 'rb') as f:
    top_keys = pickle.load(f)
    

with open('../../../Desktop/top_keys.pkl', 'rb') as f:
    top_keys_old = pickle.load(f)
```

```python
test_csv = pd.read_csv('../../../Desktop/test-balanced.csv')
```

```python
test_keys = list(test_csv['key'].values)
```

```python
by_v = {}

for i in top_keys:
    v = i[:5]
    if v not in by_v:
        by_v[v] = []
        
    by_v[v].append(i)
```

```python
tot = 0
cor = 0
for cnt, i in enumerate(hcrn_preds):
    if cnt % 100000 == 0:
        print(cnt)
    key = i['csv_q_id']
    v = key[:5]
    if key not in by_v[v]:
        continue
    if i['answer'] == i['prediction']:
        tot += 1
    cor += 1
```

```python
print("Total questions: %s" % cor)
print("Correct questions: %s" % tot)
print("Accuracy: %s" % round((tot / cor) * 100, 2))
```
