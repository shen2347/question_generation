```python
import json
import pandas as pd
```

```python
train_csv = pd.read_csv('../../../Desktop/train-balanced.csv')
val_csv = pd.read_csv('../../../Desktop/valid-balanced.csv')

csv = pd.concat([train_csv, val_csv])
```

```python
keys = list(csv['key'].values)
len(keys)

# step 1: put into by video
by_v = {}

for k in keys:
    v_id = k[:5]
    if v_id not in by_v:
        by_v[v_id] = []
    by_v[v_id].append(k)
```

```python
for v in by_v:
    
    with open('../exports/dataset/balanced/train-before/%s.txt' % v, 'r') as f:
        QA = json.load(f)
        
        
    valid_keys = by_v[v]
    
    new_QA = {}
    
    for key in valid_keys:
        new_QA[key] = QA[key]
        
        
    if len(QA) != len(new_QA):
        print(v, len(QA), len(new_QA))
    
    with open('../exports/dataset/balanced/train/%s.txt' % v, 'w+') as f:
        json.dump(new_QA, f)
```
